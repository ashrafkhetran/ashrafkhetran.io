```python
# Import libraries - "Masale taiyar hain!"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style('darkgrid')
plt.style.use('seaborn-v0_8-darkgrid')

# Load flights dataset
df = sns.load_dataset('flights')
print("âœ… Flights dataset loaded! *Bismillah, ab shuru karte hain!*")
print(f"Shape: {df.shape[0]} rows aur {df.shape[1]} columns")
```

```python
# CORRECTED Feature Engineering - "String month ka problem solve karo!"
# First convert month names to numbers
df['month_num'] = pd.to_datetime(df['month'], format='%b').dt.month

# FIXED: Use month_num (numbers) instead of month (strings)
df['year_month'] = pd.to_datetime({
    'year': df['year'],
    'month': df['month_num'],
    'day': 1
})

# Other features
df['quarter'] = ((df['month_num']-1)//3) + 1
df['season'] = df['month_num'].map({1:'Winter', 2:'Winter', 3:'Spring', 4:'Spring',
                                     5:'Spring', 6:'Summer', 7:'Summer', 8:'Summer',
                                     9:'Fall', 10:'Fall', 11:'Fall', 12:'Winter'})

print("âœ… Feature engineering successful! *Ab chalega sab theek!*")
display(df[['year', 'month', 'month_num', 'year_month', 'quarter', 'season', 'passengers']].head())
```

# 2. Data Distribution

```python
# Overall Distribution
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

# Histogram
axes[0, 0].hist(df['passengers'], bins=20, edgecolor='black', color='skyblue', alpha=0.7)
axes[0, 0].set_title('Passengers Distribution (1000s)', fontweight='bold')
axes[0, 0].axvline(df['passengers'].mean(), color='red', linestyle='--', 
                   label=f"Mean: {df['passengers'].mean():.0f}")
axes[0, 0].axvline(df['passengers'].median(), color='green', linestyle='-',
                   label=f"Median: {df['passengers'].median():.0f}")
axes[0, 0].legend()

# KDE Plot
sns.kdeplot(data=df, x='passengers', ax=axes[0, 1], fill=True, color='lightcoral')
axes[0, 1].set_title('Density Curve', fontweight='bold')

# Box Plot
axes[1, 0].boxplot(df['passengers'], patch_artist=True)
axes[1, 0].set_title('Box Plot - Outliers?', fontweight='bold')

# Year-wise distribution
sns.boxplot(x='year', y='passengers', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Distribution by Year', fontweight='bold')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.suptitle('DISTRIBUTION ANALYSIS - PASSENGERS KA SCENE!', 
             fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Monthly Distribution - "Har month ki kahani"
plt.figure(figsize=(14, 6))
monthly_stats = df.groupby('month')['passengers'].agg(['mean', 'std', 'min', 'max']).reset_index()
monthly_stats = monthly_stats.sort_values('mean', ascending=True)

# Create horizontal bar plot
plt.barh(monthly_stats['month'], monthly_stats['mean'], 
         xerr=monthly_stats['std'], capsize=5, color='lightblue', alpha=0.7)
plt.xlabel('Average Passengers (1000s)', fontweight='bold')
plt.title('MONTHLY AVERAGE PASSENGERS\n*July kyun sab se zyada hai? Summer holidays!*', 
          fontweight='bold')
plt.grid(axis='x', alpha=0.3)

# Add values on bars
for i, (month, mean) in enumerate(zip(monthly_stats['month'], monthly_stats['mean'])):
    plt.text(mean + 10, i, f'{mean:.0f}', va='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("ðŸ” *INSIGHT*: July-August mein sab se zyada passengers - *summer vacation season!*")
```

# 3. Data Relationships

```python
# Sort by date for time series analysis
df_sorted = df.sort_values(['year', 'month_num']).reset_index(drop=True)

# Trend plot
fig = px.line(df_sorted, x='year_month', y='passengers',
              title='PASSENGERS OVER TIME - 12 SAAL KA SAFAR!',
              labels={'passengers': 'Passengers (1000s)', 'year_month': 'Year'},
              hover_data=['month', 'year'])

fig.update_traces(line=dict(width=3))
fig.add_annotation(x=df_sorted.loc[df_sorted['passengers'].idxmax(), 'year_month'], 
                   y=df_sorted['passengers'].max(),
                   text=f"Peak: {df_sorted['passengers'].max():.0f}k<br>{df_sorted.loc[df_sorted['passengers'].idxmax(), 'month']} {df_sorted.loc[df_sorted['passengers'].idxmax(), 'year']}",
                   showarrow=True, arrowhead=2, arrowcolor="red")

fig.update_layout(title_font_size=22, height=500)
fig.show()
```

```python
# Scatter plots with trend lines
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Year vs Passengers (Trend)
axes[0].scatter(df['year'], df['passengers'], alpha=0.6, s=80, 
                c=df['month_num'], cmap='viridis')
axes[0].set_xlabel('Year', fontweight='bold')
axes[0].set_ylabel('Passengers (1000s)', fontweight='bold')
axes[0].set_title('Year vs Passengers\n*Trend bohat clear hai!*', fontweight='bold')

# Add trend line
z = np.polyfit(df['year'], df['passengers'], 1)
p = np.poly1d(z)
axes[0].plot(df['year'], p(df['year']), "r--", linewidth=2)

# Month vs Passengers (Seasonality)
monthly_avg = df.groupby('month_num')['passengers'].mean()
axes[1].plot(monthly_avg.index, monthly_avg.values, 'o-', 
             markersize=8, linewidth=2, color='darkorange')
axes[1].set_xlabel('Month Number', fontweight='bold')
axes[1].set_ylabel('Average Passengers', fontweight='bold')
axes[1].set_title('Month vs Passengers\n*Seasonality ka pattern!*', fontweight='bold')
axes[1].grid(alpha=0.3)

# Add month names
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
axes[1].set_xticks(range(1, 13))
axes[1].set_xticklabels(month_names)

plt.suptitle('KEY RELATIONSHIPS - TREND & SEASONALITY!', 
             fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Lag Features - "Pichhla mahina kitna tha?"
df_sorted = df.sort_values('year_month').reset_index(drop=True)
df_sorted['passengers_lag1'] = df_sorted['passengers'].shift(1)
df_sorted['passengers_lag12'] = df_sorted['passengers'].shift(12)  # Last year same month

# Autocorrelation
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Lag 1 plot
axes[0].scatter(df_sorted['passengers_lag1'], df_sorted['passengers'], 
                alpha=0.6, s=80, color='lightblue', edgecolors='black')
axes[0].set_xlabel('Previous Month Passengers', fontweight='bold')
axes[0].set_ylabel('Current Month Passengers', fontweight='bold')
axes[0].set_title('Autocorrelation (Lag 1)\n*Pichhle month kitne the?*', fontweight='bold')

z1 = np.polyfit(df_sorted['passengers_lag1'][1:], df_sorted['passengers'][1:], 1)
p1 = np.poly1d(z1)
axes[0].plot(df_sorted['passengers_lag1'], p1(df_sorted['passengers_lag1']), "r--")

# Lag 12 plot (seasonal)
axes[1].scatter(df_sorted['passengers_lag12'], df_sorted['passengers'], 
                alpha=0.6, s=80, color='lightgreen', edgecolors='black')
axes[1].set_xlabel('Last Year Same Month', fontweight='bold')
axes[1].set_ylabel('Current Month Passengers', fontweight='bold')
axes[1].set_title('Autocorrelation (Lag 12)\n*Pichhle saal isi month kitne the?*', 
                  fontweight='bold')

z12 = np.polyfit(df_sorted['passengers_lag12'][12:], df_sorted['passengers'][12:], 1)
p12 = np.poly1d(z12)
axes[1].plot(df_sorted['passengers_lag12'], p12(df_sorted['passengers_lag12']), "r--")

plt.suptitle('AUTOCORRELATION - TIME SERIES KA DNA!', 
             fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()

print(f"ðŸ” Lag-1 Correlation: {df_sorted['passengers'].corr(df_sorted['passengers_lag1']):.3f}")
print(f"ðŸ” Lag-12 Correlation: {df_sorted['passengers'].corr(df_sorted['passengers_lag12']):.3f}")
```

```python
# Rolling Statistics - "Moving average ka kamaal"
df_sorted['rolling_mean_3'] = df_sorted['passengers'].rolling(window=3).mean()
df_sorted['rolling_mean_12'] = df_sorted['passengers'].rolling(window=12).mean()

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_sorted['year_month'], y=df_sorted['passengers'],
                         mode='lines+markers', name='Actual Passengers',
                         line=dict(color='blue', width=2)))
fig.add_trace(go.Scatter(x=df_sorted['year_month'], y=df_sorted['rolling_mean_3'],
                         mode='lines', name='3-Month MA',
                         line=dict(color='red', width=1, dash='dash')))
fig.add_trace(go.Scatter(x=df_sorted['year_month'], y=df_sorted['rolling_mean_12'],
                         mode='lines', name='12-Month MA',
                         line=dict(color='green', width=2)))

fig.update_layout(title='PASSENGERS WITH ROLLING AVERAGES - SMOOTHING EFFECT!',
                  xaxis_title='Year', yaxis_title='Passengers (1000s)',
                  title_font_size=20, height=500)
fig.show()
```

# 4. Data Comparison

```python
# PILLAR 4: Data Comparison
# Monthly ranking
monthly_avg = df.groupby('month')['passengers'].mean().sort_values(ascending=False)

print("=== MONTHLY RANKING - TOP 5 BUSY MONTHS ===")
for idx, (month, avg) in enumerate(monthly_avg.head().items(), 1):
    print(f"{idx}. {month}: {avg:.0f}k passengers")

# Heatmap comparison - FIXED SYNTAX!
pivot_data = df.pivot(index="month", columns="year", values="passengers")
plt.figure(figsize=(14, 6))
sns.heatmap(pivot_data, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=0.5)
plt.title('PASSENGERS HEATMAP - MONTH vs YEAR\n*Garmi mein garmi dikhti hai!*', 
          fontweight='bold', fontsize=16)
plt.xlabel('Year', fontweight='bold')
plt.ylabel('Month', fontweight='bold')
plt.tight_layout()
plt.show()

print("ðŸ† *WINNER*: July-August (summer holidays)")
print("ðŸ“‰ *LOSER*: November (post-holiday slowdown)")
```

```python
# Yearly Comparison - "Saalo ka hisaab"
yearly_comparison = df.groupby('year').agg({
    'passengers': ['mean', 'sum', 'std']
}).round(2)
yearly_comparison.columns = ['Avg_Monthly', 'Total_Year', 'Std_Dev']

print("=== YEARLY GROWTH - KITNI TARAKKI HUI? ===")
display(yearly_comparison)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Total passengers by year
axes[0].plot(yearly_comparison.index, yearly_comparison['Total_Year'], 
             'o-', markersize=8, linewidth=3, color='darkorange')
axes[0].set_xlabel('Year', fontweight='bold')
axes[0].set_ylabel('Total Passengers (1000s)', fontweight='bold')
axes[0].set_title('Yearly Total Passengers\n*Growth ka trend dekho!*', fontweight='bold')
axes[0].grid(alpha=0.3)

# Growth rate
yearly_comparison['growth_rate'] = yearly_comparison['Total_Year'].pct_change() * 100
axes[1].bar(yearly_comparison.index[1:], yearly_comparison['growth_rate'][1:], 
            color='lightgreen', edgecolor='black')
axes[1].set_xlabel('Year', fontweight='bold')
axes[1].set_ylabel('Growth Rate (%)', fontweight='bold')
axes[1].set_title('Year-over-Year Growth Rate', fontweight='bold')
axes[1].axhline(0, color='red', linestyle='--')

plt.suptitle('YEARLY COMPARISON - SAAL DAR SAAL PROGRESS!', 
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

avg_growth = yearly_comparison['growth_rate'].mean()
print(f"\nðŸ“ˆ Average Annual Growth Rate: {avg_growth:.2f}%")
print(f"ðŸ’° *Matlab*: Business almost {avg_growth:.0f}% har saal grow kar raha hai!")
```

```python
# Seasonal Comparison - "Mausam ka asar"
seasonal_stats = df.groupby('season')['passengers'].agg(['mean', 'count']).round(2)
seasonal_stats['percentage'] = (seasonal_stats['count'] / seasonal_stats['count'].sum()) * 100

print("=== SEASONAL COMPARISON - MAUSAM BOLTA HAI! ===")
display(seasonal_stats)

# Radar chart for seasons
season_monthly = df.groupby(['season', 'month_num'])['passengers'].mean().reset_index()
season_pivot = season_monthly.pivot(index='month_num', columns='season', values='passengers')

fig = go.Figure()

colors = ['blue', 'green', 'orange', 'red']
for idx, season in enumerate(['Spring', 'Summer', 'Fall', 'Winter']):
    if season in season_pivot.columns:
        fig.add_trace(go.Scatterpolar(
            r=season_pivot[season].values,
            theta=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                   'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
            fill='toself', name=season, 
            line_color=colors[idx]
        ))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, season_pivot.max().max()])),
    title='SEASONAL PATTERNS - RADAR VIEW!\n*Summer ka peak sab ko dikhta hai!*',
    title_font_size=18, height=600
)
fig.show()
```

```python
# Quarter-wise Comparison with Statistical Tests
print("=== QUARTERLY COMPARISON - TRIMESTER ANALYSIS ===")
quarterly_stats = df.groupby('quarter')['passengers'].agg(['mean', 'std', 'count'])
quarterly_stats['std_error'] = quarterly_stats['std'] / np.sqrt(quarterly_stats['count'])
display(quarterly_stats)

# ANOVA test
from scipy import stats
q1_data = df[df['quarter']==1]['passengers'].values
q2_data = df[df['quarter']==2]['passengers'].values
q3_data = df[df['quarter']==3]['passengers'].values
q4_data = df[df['quarter']==4]['passengers'].values

f_stat, p_value = stats.f_oneway(q1_data, q2_data, q3_data, q4_data)
print(f"\nðŸ”¬ ANOVA Test: F-statistic = {f_stat:.3f}, p-value = {p_value:.2e}")
if p_value < 0.05:
    print("âœ… *Bohat significant difference hai quarters mein!*")
else:
    print("âŒ Koi significant difference nahi hai")

# Box plot with quarters
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x='quarter', y='passengers', data=df, ax=ax, palette='Set2')
ax.set_xlabel('Quarter', fontweight='bold')
ax.set_ylabel('Passengers (1000s)', fontweight='bold')
ax.set_title('QUARTERLY COMPARISON - Q3 SAB SE AAGE!\n*July-September ka magic*', 
             fontweight='bold')

# Add statistics on plot
for quarter in [1, 2, 3, 4]:
    mean_val = df[df['quarter']==quarter]['passengers'].mean()
    ax.text(quarter-1, mean_val + 30, f'{mean_val:.0f}', 
            ha='center', fontweight='bold', fontsize=12)

plt.tight_layout()
plt.show()
```
