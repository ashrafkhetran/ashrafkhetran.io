# 1. Introduction
The Iris dataset is one of the most famous datasets in machine learning and statistics. It contains measurements of iris flowers across three species (setosa, versicolor, virginica). This notebook explores the dataset through composition, distribution, comparison, and relationships.

# 2. Load data

```python
# 2 Load Data
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Theme
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams["figure.dpi"] = 120

# Load iris dataset
iris = sns.load_dataset("iris")

# Basic info
print(iris.shape)
iris.head()
```

# 3. Composition

```python
# Count of species
species_counts = iris["species"].value_counts()

plt.figure(figsize=(6,4))
sns.countplot(data=iris, x="species", palette="Set2")
plt.title("Composition of Iris Dataset by Species", fontsize=12, weight="bold")
plt.show()

species_counts
```

Interpretation:

The dataset is balanced: 50 samples each of setosa, versicolor, and virginica.

This makes it ideal for classification tasks.

# 4. Distribution

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.histplot(iris["sepal_length"], kde=True, ax=axes[0,0], color="teal")
axes[0,0].set_title("Distribution of Sepal Length")

sns.histplot(iris["sepal_width"], kde=True, ax=axes[0,1], color="orange")
axes[0,1].set_title("Distribution of Sepal Width")

sns.histplot(iris["petal_length"], kde=True, ax=axes[1,0], color="purple")
axes[1,0].set_title("Distribution of Petal Length")

sns.histplot(iris["petal_width"], kde=True, ax=axes[1,1], color="green")
axes[1,1].set_title("Distribution of Petal Width")

plt.tight_layout()
plt.show()
```

Interpretation:

Sepal measurements are more normally distributed.

Petal measurements show clear separation between species, especially petal length.

# 5. Comparison

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.boxplot(data=iris, x="species", y="sepal_length", ax=axes[0,0], palette="Set2")
axes[0,0].set_title("Sepal Length by Species")

sns.boxplot(data=iris, x="species", y="sepal_width", ax=axes[0,1], palette="Set2")
axes[0,1].set_title("Sepal Width by Species")

sns.boxplot(data=iris, x="species", y="petal_length", ax=axes[1,0], palette="Set2")
axes[1,0].set_title("Petal Length by Species")

sns.boxplot(data=iris, x="species", y="petal_width", ax=axes[1,1], palette="Set2")
axes[1,1].set_title("Petal Width by Species")

plt.tight_layout()
plt.show()
```

Interpretation:

Setosa is clearly distinct with small petals.

Versicolor and Virginica overlap more, but Virginica tends to have larger petals.

Sepal width varies less across species compared to petal features.

# 6. Relationships

```python
# Pairplot for relationships
sns.pairplot(iris, hue="species", palette="Set2", diag_kind="kde")
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02, fontsize=14, weight="bold")
plt.show()

# Correlation heatmap
plt.figure(figsize=(6,5))
sns.heatmap(iris.drop(columns="species").corr(), annot=True, cmap="Blues")
plt.title("Correlation Matrix of Iris Features", fontsize=12, weight="bold")
plt.show()
```

Interpretation:

Petal length and petal width are highly correlated.

Sepal features are less correlated with petal features.

Pairplots show clear clustering by species, especially in petal dimensions.

# 7. Summary
Composition: Balanced dataset with 3 species, 50 samples each.

Distribution: Petal features show distinct separation; sepal features overlap more.

Comparison: Boxplots highlight species differences, especially in petal length/width.

Relationships: Strong correlation between petal length and width; species clusters visible in pairplots.
