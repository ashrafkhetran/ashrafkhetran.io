# 1. Introduction
The Palmer Penguins dataset is a modern alternative to the Iris dataset. It contains measurements of penguins across three species (Adelie, Chinstrap, Gentoo) collected from the Palmer Archipelago in Antarctica.
This notebook explores the dataset through composition, distribution, comparison, and relationships.

# 2. Load Data

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Theme
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams["figure.dpi"] = 120

# Load penguins dataset
penguins = sns.load_dataset("penguins")

# Basic info
print(penguins.shape)
penguins.head()
```

# 3. Composition

```python
# Count of species
species_counts = penguins["species"].value_counts()

plt.figure(figsize=(6,4))
sns.countplot(data=penguins, x="species", palette="Set2")
plt.title("Composition of Penguins Dataset by Species", fontsize=12, weight="bold")
plt.show()

species_counts
```

# 4. Distribution

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.histplot(penguins["bill_length_mm"].dropna(), kde=True, ax=axes[0,0], color="teal")
axes[0,0].set_title("Distribution of Bill Length (mm)")

sns.histplot(penguins["bill_depth_mm"].dropna(), kde=True, ax=axes[0,1], color="orange")
axes[0,1].set_title("Distribution of Bill Depth (mm)")

sns.histplot(penguins["flipper_length_mm"].dropna(), kde=True, ax=axes[1,0], color="purple")
axes[1,0].set_title("Distribution of Flipper Length (mm)")

sns.histplot(penguins["body_mass_g"].dropna(), kde=True, ax=axes[1,1], color="green")
axes[1,1].set_title("Distribution of Body Mass (g)")

plt.tight_layout()
plt.show()
```

# 5. Comparison

```python
fig, axes = plt.subplots(2, 2, figsize=(12,8))

sns.boxplot(data=penguins, x="species", y="bill_length_mm", ax=axes[0,0], palette="Set2")
axes[0,0].set_title("Bill Length by Species")

sns.boxplot(data=penguins, x="species", y="bill_depth_mm", ax=axes[0,1], palette="Set2")
axes[0,1].set_title("Bill Depth by Species")

sns.boxplot(data=penguins, x="species", y="flipper_length_mm", ax=axes[1,0], palette="Set2")
axes[1,0].set_title("Flipper Length by Species")

sns.boxplot(data=penguins, x="species", y="body_mass_g", ax=axes[1,1], palette="Set2")
axes[1,1].set_title("Body Mass by Species")

plt.tight_layout()
plt.show()
```

# 6. Relationships

```python
# Pairplot for relationships
sns.pairplot(penguins.dropna(), hue="species", palette="Set2", diag_kind="kde")
plt.suptitle("Pairwise Relationships in Penguins Dataset", y=1.02, fontsize=14, weight="bold")
plt.show()

# Correlation heatmap
plt.figure(figsize=(6,5))
sns.heatmap(penguins.drop(columns=["species","island","sex"]).corr(), annot=True, cmap="Blues")
plt.title("Correlation Matrix of Penguin Features", fontsize=12, weight="bold")
plt.show()
```
