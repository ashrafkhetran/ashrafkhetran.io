```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
```

```python
df = sns.load_dataset('tips')
# Keep numerical columns and one categorical column
df_subset = df[['total_bill', 'tip', 'size', 'sex']]
plt.figure(figsize=(10,6))
parallel_coordinates(df_subset, 'sex', colormap=plt.cm.Set1)
plt.title("Parallel Coordinates Plot - Tips Dataset")
plt.xlabel("Features")
plt.ylabel("Values")
plt.show()
```

```python
parallel_coordinates(df[['total_bill','tip','size','day']], 'day', colormap=plt.cm.Set2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[['total_bill','tip','size']]), columns=['total_bill','tip','size'])
df_scaled['sex'] = df['sex']
parallel_coordinates(df_scaled, 'sex', colormap=plt.cm.Set1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
df = sns.load_dataset('tips')

# select categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# plot pie chart for each categorical column
for col in categorical_cols:
    counts = df[col].value_counts()
    plt.figure(figsize=(6,6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('tips')

# Create histograms side by side
plt.figure(figsize=(12,5))

# Histogram for total_bill
plt.subplot(1,2,1)
plt.hist(df['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

# Histogram for tip
plt.subplot(1,2,2)
plt.hist(df['tip'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Tips')
plt.xlabel('Tip ($)')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('tips')

categorical_cols = ['sex', 'smoker', 'day', 'time']

plt.figure(figsize=(14,10))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(2,2,i)
    sns.barplot(x=col, y='total_bill', data=df, estimator='mean', ci=None, palette='Set2')
    plt.title(f"Average Total Bill by {col}")
    plt.ylabel("Average Total Bill ($)")
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('tips')

# Add 'size' to categorical columns list
categorical_cols = ['sex', 'smoker', 'day', 'time', 'size']

plt.figure(figsize=(16,12))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(3,2,i)
    sns.barplot(x=col, y='total_bill', data=df, estimator='mean', ci=None, palette='Set2')
    plt.title(f"Average Total Bill by {col}")
    plt.ylabel("Average Total Bill ($)")
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('tips')

# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap - Tips Dataset (Numeric Only)")
plt.show()
```
