# 1. Introduction
The Titanic dataset is a classic dataset used for statistical learning and machine learning. It contains information about passengers aboard the Titanic, including demographics, class, fare, and survival status.
This notebook explores the dataset through composition, distribution, comparison, and relationships.

# 2. Load Data

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Theme
sns.set_theme(style="whitegrid", context="notebook")
plt.rcParams["figure.dpi"] = 120

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Basic info
print(titanic.shape)
titanic.head()
```

# 3. Composition

```python
# Count survival
survival_counts = titanic["survived"].value_counts()

plt.figure(figsize=(6,4))
sns.countplot(data=titanic, x="survived", palette="Set2")
plt.title("Composition of Titanic Dataset by Survival", fontsize=12, weight="bold")
plt.xticks([0,1], ["Did not survive", "Survived"])
plt.show()

survival_counts
```

Interpretation:

About 38% survived, while 62% did not.

This imbalance is important for predictive modeling.

# 4. Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.histplot(titanic["age"].dropna(), kde=True, bins=30, ax=axes[0], color="teal")
axes[0].set_title("Distribution of Passenger Age")

sns.histplot(titanic["fare"].dropna(), kde=True, bins=30, ax=axes[1], color="orange")
axes[1].set_title("Distribution of Passenger Fare")

plt.tight_layout()
plt.show()
```

Interpretation:

Age distribution is rightskewed, with many young passengers.

Fare distribution is highly skewed, with most fares low but some very high outliers.

# 5. Comparison

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.boxplot(data=titanic, x="class", y="age", ax=axes[0], palette="Set2")
axes[0].set_title("Age Distribution by Class")

sns.boxplot(data=titanic, x="class", y="fare", ax=axes[1], palette="Set2")
axes[1].set_title("Fare Distribution by Class")

plt.tight_layout()
plt.show()
```

Interpretation:

Firstclass passengers tended to pay much higher fares.

Age distribution is fairly similar across classes, though variability exists.

# 6. Relationships

```python
# Survival by class and sex
plt.figure(figsize=(8,5))
sns.barplot(data=titanic, x="class", y="survived", hue="sex", ci=None, palette="Set2")
plt.title("Survival Rate by Class and Sex", fontsize=12, weight="bold")
plt.show()

# Pairplot for numeric features
sns.pairplot(titanic[["age","fare","survived","pclass"]].dropna(), hue="survived", palette="Set2")
plt.suptitle("Pairwise Relationships in Titanic Dataset", y=1.02, fontsize=14, weight="bold")
plt.show()
```

Interpretation:

Women had higher survival rates across all classes.

Firstclass passengers had better survival chances compared to thirdclass.

Pairplots show survival clustering with higher fares and younger ages.

# 7. Summary
Composition: 62% did not survive, 38% survived.

Distribution: Age skewed toward younger passengers; fares skewed with high outliers.

Comparison: Firstclass fares much higher; age fairly similar across classes.

Relationships: Survival strongly linked to sex and class; higher fares and younger ages correlated with survival.
